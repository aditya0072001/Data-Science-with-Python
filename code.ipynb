Operators

```
# Examples of Arithmetic Operator

# Addition of numbers
add = a + b

# Subtraction of numbers
sub = a - b

# Multiplication of number
mul = a * b

# Division(float) of number
div1 = a / b

# Division(floor) of number
div2 = a // b

# Modulo of both number
mod = a % b

# Power
p = a ** b

# print results
print(add)
print(sub)
print(mul)
print(div1)
print(div2)
print(mod)
print(p)
```

```
# Examples of Relational Operators

# a > b is False
print(a > b)

# a < b is True
print(a < b)

# a == b is False
print(a == b)

# a != b is True
print(a != b)

# a >= b is False
print(a >= b)

# a <= b is True
print(a <= b)
```

```
# Examples of Logical Operator
a1 = True
b1 = False

# Print a and b is False
print(a1 and b1)

# Print a or b is True
print(a1 or b1)

# Print not a is False
print(not a1)
```

```
# Examples of Bitwise operators

# Print bitwise AND operation
print(a & b)

# Print bitwise OR operation
print(a | b)

# Print bitwise NOT operation
print(~a)

# print bitwise XOR operation
print(a ^ b)

# print bitwise right shift operation
print(a >> 2)

# print bitwise left shift operation
print(a << 2)
```

```
# Examples of Assignment Operators

# Assign value
b = a
print(b)

# Add and assign value
b += a
print(b)

# Subtract and assign value
b -= a
print(b)

# multiply and assign
b *= a
print(b)

# bitwise lishift operator
b <<= a
print(b)
```

### Python Condition Programming

Discussing about various conditional programming concepts present in python such as OOPS, conditional statements and  exception handling

```
# Loops

# while loop

i = 0

print("printing while loop")

while (i < 5):
    print(i)
    i = i +1

# for loop

print("printing for loop")

for i in range(0,5):
    print(i)

# loop control statements

print("continue control statment")

for letter in "this_is_myarticle":
    if letter == "e":
        continue  # continue returns control to the beginning of the loop
 
    print("Letter currently,",letter)

print("break control statment")

for letter in "this_is_myarticle":
    if letter == "e":
        break  # break stop the control of loop
 
    print("Letter currently,",letter)

print("pass control statment")

for letter in "this_is_myarticle":
    pass # empty loop kinda like place holder

print("Letter currently,",letter)

# loop with else

print("Loop with else statment")

for i in range(0,3):
    print(i)
else:
    print("if no break statment in loop else executes")
```

```
# OOP's Object Oriented Programming uses classes and objects it provides real life entities like inheritance, polymorphisms, encapsulations and more.

# Class - collection of objects , it contains blueprint for objects throught which objects are made.

class Human:
    pass  # classes are made with class keyword, attributes are variables that belong to class, attributes are always public and can be accessed using dot operator.

# e.g Human.leftarm

# Object - entity with state and behaviour along with it

child_object = Human()

# self is similar to pointer in C++, the class method must have extra first parameter in method definition.

# __init__ is similar to constructor in C++ and Java and seen as soon as object of a class is instantiated.


class Human:
    cattr = "mammal"

    def __init__(self,name):
        self.name = name


aditya = Human("Aditya") # object creation

# accessing class attributes
print(aditya.__class__.cattr)

# accessing class instance
print(aditya.name)

# Inheritance - using/inhertiing properities from another class

class Dog(object): # parent class
	def __init__(self, name):
		self.name = name
	
class Breed(Dog): # child class
	def __init__(self, name):
		Dog.__init__(self, name)
		
	def detail(self):
		print("Dog Breed is {}".format(self.name))


# creation of an object variable or an instance
a = Breed('German Shephard')

a.detail()

# Polymorphism - having many forms

class Dog:
    def detail(self):
        print("this is dog class method")

class Animal(Dog):
    def detail(self):
        print("this is animal class method")

obj_dog = Dog()
obj_animal = Animal()

print(obj_dog.detail(),obj_animal.detail())

# Encapsulation - it wraps data and methods in one unit

class Animal: # class is a example as it contains methods , variables etc
    category = ""
    def __init__(self,name):
        self.name = name
    
class Insect(Animal):
    def __init__(self):
        Animal.__init__(self)
        Animal.category = "insect"
        print(self.category)

a = Animal("Beatle")
print(a.name)
```

```
# function - block of staments that return the specific task

# function definition

def fun(a): # def is used to define function and fun is function name and a is parameter
    print("Number supplied to function fun is",a)

fun(10) # calling the function

def sum(a,b):
    return a + b # return statment returns the value when the function is called

print(sum(3,2)) # this function returns sum of 2 numbers that are supplied

def pname(name=None):
    print("Name provided to function pname is",name)

pname() # calling the function with default parameters

# variable length arguments *argv - non keyword attributes and **kwargs - keyword attributes with key and value pair

def non_keypair(*argv):
    for variable in argv:
        print(variable)

def keypair(**kwargs):
    for key,value in kwargs.items():
        print(key,value)

non_keypair("hello","world") # non keypair function
keypair(f1="hello",f2="world") # keypair function

# pass by reference 

def xlist(x):
    x[2] = 45

l1 = [1,2,3,4]
xlist(l1)
print(l1)

# pass by value

def ylist(y):
    y = [10,20,30]

ylist(l1)
print(l1)

# anonymous function fnctions that are not named

square = lambda x : x*x

print(square(2))
```

Exception handling in python is managed using three keywords try, except and finally. Exceptions are instances that are raised when some internal event occurs which changes the flow of program.

```
# exception example

value = 10

if value > 9:
print("its bigger than 9")
```

```
# handling the the below error of divison by zero

value = 10

try:
    if value / 0:
        print("dividing value varriable by zero")

except:
    print("error occured")

# can specify specific error like 

try:
    if value / 0:
        print("dividing value varriable by zero")

except ZeroDivisionError:

    print("ZeroDivisionError occured")

finally:

    print("finally executes regarless or whether exception occured or not")

# Raising own exception with raise keyword

try:
    raise("Raising exception")
    print("Try block")
except:
    print("Exception occured")
```

### Python Data Manipulation 

Data manipulation techniques and ways present in Python for Data Science work namely data preprocessing, data transformation and data visualization.

Data preprocessing is the task of converting data from a given form to a much useful and consumable form.

```
# performing array operation using numpy

import numpy as np

a1 = np.arange(4,dtype=np.float_).reshape(2,2)

print(a1)

a2 = np.array([10,20])

print(a2)

# array operations - add subtract multiply divide

print(np.add(a1,a2))

print(np.subtract(a1,a2))

print(np.multiply(a1,a2))

print(np.divide(a1,a2))
```

Data Transformation is process of converting data from one format or structure into another format or structure

```
# Pandas - This library will be used for transforming dataframes using ways such as slicing,indexing, manipualting and cleaning

import pandas as pd

employee_list = [
    ["Aditya",21,4500000],
    ["Annie",22,3400000],
    ["Aman",25,1200000]
]

df = pd.DataFrame(employee_list, columns = ['Name','Age','Salary'])

print(df)

# slicing columns in dataframe

print(df.iloc[:,0:2])

# adding new column derived from other column(s)

df["Tax"] = df["Salary"]*0.45

print(df)

# sorting dataframe

df.sort_values(by=['Tax'])

print(df)

# checking for missing values

print(df.isnull().sum())

# df = df.dropna() in case of nan values and missing values

print(df.isnull()) # returns True if null else false

# filling null value with values

df.fillna(0) # for filling with zero

df.fillna(method='pad') # filling previous value

df.fillna('Empty', inplace = True) # filling na value with "Empty" string value


# replacing NaN value in dataframe with 1 value

df.replace(to_replace = np.nan , value = "1")

# reading csv file in pandas is done using pd.read_csv("path_to_file")

# reading json file in pandas is done using pd.read_json("path_to_file")
```

Data Visualization is the process of presenting data in the form of graphs or charts. It helps to understand large and complex amounts of data very easily. It allows the decision-makers to make decisions very efficiently and also allows them in identifying new trends and patterns very easily.

```
# data visualization using matplot library

import matplotlib.pyplot as plt

x = [10, 20]
y = [20, 25]
 
# plotting the data
plt.plot(x, y)

plt.title("Straight Line") # adding title

plt.ylabel("Y axis") # Y axis label
plt.xlabel("X axis") # X axis label

plt.show()
```

```
plt.plot(x, y)
 
# Adding title to the plot
plt.title("Linear graph", fontsize=20, color="Blue")
 
# Adding label on the y-axis
plt.ylabel('Y-Axis')
 
# Adding label on the x-axis
plt.xlabel('X-Axis')
 
# Setting the limit of y-axis
plt.ylim(0, 80)
 
# setting the labels of x-axis
plt.xticks(x, labels=["one", "two"])

# Adding legends
plt.legend(["GFG"])
 
plt.show()
```

```
from matplotlib.figure import Figure

fig = plt.figure(figsize = (5, 4))

# Adding the axes to the figure
ax = fig.add_axes([1, 1, 1, 1])

# plotting 1st dataset to the figure
ax1 = ax.plot(x, y)

# plotting 2nd dataset to the figure
ax2 = ax.plot(y, x)

# Setting Title
ax.set_title("Linear Graph")

# Setting Label
ax.set_xlabel("X-Axis")
ax.set_ylabel("Y-Axis")

# Adding Legend
ax.legend(labels = ('line 1', 'line 2'))

plt.show()
```

```
# sub plot in matplot

# Creating figure object
plt.figure()

# adding first subplot
plt.subplot(121)
plt.plot(x, y)

# adding second subplot
plt.subplot(122)
plt.plot(y, x)
```
